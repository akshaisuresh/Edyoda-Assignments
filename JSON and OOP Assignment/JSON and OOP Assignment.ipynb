## Assignment 1

```python
import json

class Employee:
    def __init__(self, name, dob, height, city, state):
        self.name = name
        self.dob = dob
        self.height = height
        self.city = city
        self.state = state

def read_employee_data(file_name):
    with open(file_name, 'r') as file:
        data = json.load(file)
        employees = []
        for item in data:
            name = item['Name']
            dob = item['DOB']
            height = item['Height']
            city = item['City']
            state = item['State']
            employee = Employee(name, dob, height, city, state)
            employees.append(employee)
        return employees

def save_to_json(employees, file_name):
    data = []
    for employee in employees:
        data.append({
            'Name': employee.name,
            'DOB': employee.dob,
            'Height': employee.height,
            'City': employee.city,
            'State': employee.state
        })
    with open(file_name, 'w') as file:
        json.dump(data, file, indent=4)

def print_employee_data(employees):
    for employee in employees:
        print("Name:", employee.name)
        print("DOB:", employee.dob)
        print("Height:", employee.height)
        print("City:", employee.city)
        print("State:", employee.state)
        print()

def save_states_capitals_to_json(states_capitals, file_name):
    with open(file_name, 'w') as file:
        json.dump(states_capitals, file, indent=4)

# Read employee data from JSON
employees = read_employee_data('employee.json')
print("List of Employee Objects:")
print_employee_data(employees)

# Save employee data to JSON
save_to_json(employees, 'employee_updated.json')
print("Employee data saved to employee_updated.json")

# Dictionary of states and capitals
states_capitals = {
    'Kerala': 'Thiruvananthapuram',
    'Maharashtra': 'Mumbai',
    'Tamil Nadu': 'Chennai',
    'Karnataka': 'Bengaluru',
    'Delhi': 'New Delhi',
    'Gujarat': 'Gandhinagar',
    'Rajasthan': 'Jaipur'
}

# Save states and capitals to JSON
save_states_capitals_to_json(states_capitals, 'states_capitals.json')
print("States and capitals data saved to states_capitals.json")
```

## Assignment 2

```python
class Dog:
    def __init__(self, name, age, coat_color):
        self.name = name
        self.age = age
        self.coat_color = coat_color

    def description(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

    def get_info(self):
        print(f"Coat Color: {self.coat_color}")


class JackRussellTerrier(Dog):
    def __init__(self, name, age, coat_color):
        super().__init__(name, age, coat_color)

    def action(self):
        print(f"{self.name} is running and jumping.")


class Bulldog(Dog):
    def __init__(self, name, age, coat_color):
        super().__init__(name, age, coat_color)

    def sleep(self):
        print(f"{self.name} is sleeping peacefully.")


# Create objects and implement functionalities
dogs = []

dog1 = JackRussellTerrier("Max", 3, "White and Brown")
dogs.append(dog1)

dog2 = Bulldog("Rocky", 5, "Brindle")
dogs.append(dog2)

dog3 = Bulldog("Bella", 2, "Fawn")
dogs.append(dog3)

dog4 = JackRussellTerrier("Charlie", 4, "White and Brown")
dogs.append(dog4)

dog5 = Bulldog("Lucy", 7, "Brindle")
dogs.append(dog5)

dog6 = JackRussellTerrier("Cooper", 2, "White and Brown")
dogs.append(dog6)

dog7 = Bulldog("Lola", 6, "White")
dogs.append(dog7)

dog8 = JackRussellTerrier("Daisy", 1, "White and Brown")
dogs.append(dog8)

dog9 = Bulldog("Maximus", 4, "Brindle")
dogs.append(dog9)

dog10 = JackRussellTerrier("Molly", 5, "White and Brown")
dogs.append(dog10)

# Print dog information
for dog in dogs:
    print("Dog Information:")
    dog.description()
    dog.get_info()
    if isinstance(dog, JackRussellTerrier):
        dog.action()
    elif isinstance(dog, Bulldog):
        dog.sleep()
    print()
```
